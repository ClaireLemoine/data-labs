```python
# drop object column with the IDs

CCD_norm = CCD.drop('CUST_ID', axis=1)
CCD_norm.head()
```

```python
# normalize

from sklearn import preprocessing
columns = CCD_norm.columns
x = CCD_norm.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
CCD_norm = pd.DataFrame(x_scaled, columns = columns)
CCD_norm.head()
```

```python
# check for the good number of cluster

from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,20))
visualizer.fit(CCD_norm)
visualizer.poof()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=5)
df_clusters=kmeans.fit(CCD_norm)
```

```python
# create a cluster column with the Cluster column

CCD['Cluster']=df_clusters.fit_predict(CCD_norm)
```

```python
#calculate silhouette

import sklearn.metrics as m
y_a = kmeans.predict(CCD_norm)
print("Silhouette Score: {}".format(m.silhouette_score(CCD_norm, y_a)))
```

```python
# make TSNE

from sklearn.manifold import TSNE

fig,ax=plt.subplots(1,4,figsize=(15,7))#1 row,4slides and fig size
for i,perp in enumerate([30,45,60,90]): #rotation angle
   tsne=TSNE(perplexity=perp)
   new_x=tsne.fit_transform(CCD_norm)
   ax[i].scatter(new_x[:,0],new_x[:,1],c=CCD.Cluster)#build scatter plot. X axis:0columns of numpy array, y fisrt column and all the rows. And color it
   ax[i].set_title('df perp ='+str(perp)) #set title
plt.show()
```
