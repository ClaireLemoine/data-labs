```python
#import librairy
import pandas as pd
import numpy as np
```

```python
#open file
clean_df=pd.read_csv('C:/Users/Claire/Downloads/clean_porn_df.csv')
```

```python
# set index
clean_df.set_index('Unnamed: 0', inplace=True)
```

```python
clean_df.head()
```

```python
#drop useless
clean_df.drop('Unnamed: 0.1', axis=1, inplace=True)
```

```python
#create a list of categorie columns
lst_col=[]
for i in clean_df.columns:
    if '_straight' in i:
        lst_col.append(i)
```

```python
# sum all top 10 columns

clean_df['World_straight']=clean_df[lst_col].sum(axis=1)
```

```python
# sort by sum
clean_df.sort_values("World_straight", axis = 0, ascending = False).head(10)
```

```python
# turn into list
clean_df.categories=clean_df.categories.apply(eval)
```

```python
# create a list with unique values of cat

lst_cat=[]

for i in clean_df.categories:
    for j in i:
        if j not in lst_cat:
            lst_cat.append(j)
print(lst_cat)
```

```python
# Ã  amÃ©liorer
df_with_dummies = clean_df.copy()

for i in lst_cat:
    df_with_dummies['Blowjob']=df_with_dummies.categories.apply(lambda x : np.where('Blowjob' in x,1,0))

df_with_dummies[df_with_dummies['Blowjob']==1]
```

```python
for i in lst_cat:
    df_with_dummies[i]=df_with_dummies.categories.apply(lambda x : np.where(i in x,1,0))

df_with_dummies[lst_cat]
```
