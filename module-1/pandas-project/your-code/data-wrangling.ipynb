```python
#import pandas

import pandas as pd
import re
import numpy as np
import time
```

```python
#import data

sharks = pd.read_csv('C:/Users/Claire/Downloads/GSAF5/GSAF5.csv', engine = 'python')
```

```python
#check shape and info

print(sharks.shape)
print("")
sharks.info()
```

```python
#Check for data type

sharks.dtypes
```

```python
#check for dupplicates

before = len(sharks)
before
sharks = sharks.drop_duplicates()
after = len(sharks)
after
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#check the amount of null in each column

null_cols=sharks.isnull().sum()
null_cols
```

```python
# drop the column with over 4500 null cells

null_cols[null_cols>4500]
drop_cols=list(null_cols[null_cols>4500].index)
sharks1=sharks.drop(drop_cols, axis = 1)
sharks1=sharks1.drop(['Name','Investigator or Source','pdf', 'href formula', 'href','original order', 'Injury'], axis=1)
sharks1
```

```python
# Check id Case Number = Case Number 1 = Case Number 2

sharks1[(sharks1['Case Number']==sharks1['Case Number.1']) & (sharks1['Case Number']==sharks1['Case Number.2'])]
```

```python
# Since Case Number = Case Number 1 = Case Number 2, we can remove the last 2

sharks1=sharks1.drop(['Case Number.1','Case Number.2'], axis = 1)
```

```python
#Checking if relevant
sharks1['Type'].value_counts()
```

```python
#Cheking if in columns we have similar infos but with different writings and group

print(set(sharks1['Type']))

sharks1['Type'] = sharks1['Type'].str.replace('Invalid', 'Unprovoked')
sharks1['Type'] = sharks1['Type'].str.replace('Boating', 'Unprovoked')
sharks1['Type'] = sharks1['Type'].str.replace('Boat', 'Unprovoked')
sharks1['Type'] = sharks1['Type'].str.replace('Sea Disaster', 'Unprovoked')
```

```python
# Rename columns to avoid spaces

sharks1.rename(columns={'Case Number': 'Case_Number', 'Fatal (Y/N)': 'Fatal', 'Sex ': 'Sex'}, inplace=True)
```

```python
#Checking if relevant
sharks1['Sex'].value_counts()
```

```python
# Remove all value in gender which aren't M or F
print(sharks1.Sex.unique())
sharks1['Sex'] = sharks1['Sex'].replace(np.nan, "Don't know")

sharks1['Sex'] = sharks1['Sex'].str.replace('M ', 'M')
sharks1['Sex'] = sharks1['Sex'].str.replace('lli', "Don't know")
sharks1['Sex'] = sharks1['Sex'].str.replace('N', "Don't know")
sharks1['Sex'] = sharks1['Sex'].str.replace('.', "Don't know")
```

```python
#Checking if relevant
sharks1['Fatal'].value_counts()
```

```python
print(sharks1.Fatal.unique())
sharks1.loc[(sharks1['Fatal']!='Y')&(sharks1['Fatal']!='N')] = "Don't know"
```

```python
#Clean the Age column by getting only the digits ===> to be improved

sharks1.Age = sharks1.Age.str.extract('(\d{1,2})')
```
